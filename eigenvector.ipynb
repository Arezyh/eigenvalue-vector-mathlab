```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# JUDUL APLIKASI
print("===================================================")
print("           KALKULATOR EIGENVEKTOR                ")
print("===================================================")

# PETUNJUK PENGGUNAAN APLIKASI
print("PETUNJUK PENGGUNAAN:")
print("1. MATRIKS HARUS BERBENTUK PERSEGI")
print("2. MASUKKAN UKURAN MATRIKS, MISAL 2 UNTUK 2X2; 3 UNTUK 3X3")
print("")

# FUNGSI UNTUK MEMASUKKAN NILAI KE DALAM MATRIKS
def get_matrix(size):
    matrix = []
    print(f"Masukkan nilai untuk matriks {size}x{size}:")
    for i in range(size):
        row = []
        for j in range(size):
            element = float(input(f"Masukkan nilai matriks pada posisi ({i+1}, {j+1}): "))
            row.append(element)
        matrix.append(row)
    return np.array(matrix)

# FUNGSI UNTUK MENGHITUNG EIGENVEKTOR DARI MATRIKS
def eigenvector_calculator():
    # Memasukkan ukuran matriks dari pengguna
    size = int(input("Masukkan ukuran matriks (misal, 2 untuk 2x2, 3 untuk 3x3): "))

    # Memasukkan nilai-nilai matriks dari pengguna
    print("\nMasukkan nilai matriks Anda di bawah!")
    matrix_A = get_matrix(size)

    # Memeriksa apakah matriks tersebut persegi (baris = kolom)
    if matrix_A.shape[0] != matrix_A.shape[1]:
        print("Invalid! Matriks harus persegi.")
        print("Jumlah baris dan kolom harus sama!")
        return

    # Menampilkan matriks yang dimasukkan oleh pengguna
    print("\nMatriks Anda adalah:")
    print(matrix_A)

    # Menghitung nilai eigen dan eigenvector
    eigenvalues, eigenvectors = np.linalg.eig(matrix_A)

    print("\nNilai Eigenvalues:")
    print(eigenvalues)

    print("\nNilai Eigenvector:")
    for i in range(len(eigenvalues)):
        eigenvalue = eigenvalues[i]
        eigenvector = eigenvectors[:, i]

        print(f'\n→ Eigenvector untuk nilai eigenvalue {eigenvalue}:\n{eigenvector}')

    # Plot eigenvectors secara dinamis dengan batasan plot yang fleksibel
    x = np.array([[1] for _ in range(size)])
    b = np.dot(matrix_A, x)
    max_val = max(np.max(np.abs(x)), np.max(np.abs(b))) + 1
    plot_vect(x, b, (-max_val, max_val), (-max_val, max_val))

# FUNGSI UNTUK MENGGAMBAR EIGENVEKTOR
def plot_vect(x, b, xlim, ylim):
    plt.figure(figsize=(10, 6))
    plt.quiver(0, 0, x[0], x[1], \
               color='k', angles='xy', \
               scale_units='xy', scale=1, \
               label='Vektor Asli')
    plt.quiver(0, 0, b[0], b[1], \
               color='g', angles='xy', \
               scale_units='xy', scale=1, \
               label='Vektor Transformasi')
    plt.title('Plot dari Eigenvektor')
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()

# Panggil fungsi untuk menjalankan kalkulator eigenvektor
eigenvector_calculator()
```
